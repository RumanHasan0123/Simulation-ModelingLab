```python
#Student's Assignment 1: Basic Matrix Operations

#1. Create two 3×3 matrices with random integers in Python.
Matrix_A = np.random.randint(1, 100,(3,3))
Matrix_B = np.random.randint(1, 100,(3,3))


#2. Perform addition, subtraction, and multiplication operations on these matrices.

# Addition
matrix_add= Matrix_A + Matrix_B

#Substraction
matrix_sub=Matrix_A - Matrix_B

#Multiplication
matrix_multi=Matrix_A@Matrix_B


#3. Compute the determinant, inverse, and rank of one of the matrices.

#determinant
det_A =np.linalg.det(Matrix_A)
det_B =np.linalg.det(Matrix_B)

#Inverse

#Inverse Of Matrix A
if np.linalg.det(Matrix_A) != 0:
    inv_A =np.linalg.inv(Matrix_A)
else:
    inv_A ="[***]{ERROR_01}The Matrix is Singular , Cannot Compute Inverse[***]"

#inverse Of Matrix B
if np.linalg.det(Matrix_B) != 0:
    inv_B =np.linalg.inv(Matrix_B)
else:
    inv_B ="[***]{ERROR_01}The Matrix is Singular , Cannot Compute Inverse[***]"


#Rank of Matrices

rank_Matrix_A =np.linalg.matrix_rank(Matrix_A)
rank_Matrix_B =np.linalg.matrix_rank(Matrix_B)



#Printing the output

#Print the Original Matrices
print("Matrix A:\n",Matrix_A)
print("Matrix B:\n",Matrix_B)

#Print addition , substraction and multiplication of matrices
print("Matrix Addition:\n",matrix_add)
print("Matrix Substraction:\n",matrix_sub)
print("Matrix Multiplication:\n",matrix_multi)

#print determinant , inverse and rank of matrices

print("Determinant of Matrix A\n",det_A)
print("Determinant of Matrix B\n",det_B)
print("Inverse of Matrix A\n",inv_A)
print("Inverse of Matrix B\n",inv_B)
print("Rank of Matrix A\n",rank_Matrix_A)
print("Rank of Matrix B\n",rank_Matrix_B)




```

```python
#Student's Assignment 2: Matrix Plotting and Visualization


# a. Create two vectors, each containing 15 random floats of your choice.
A = np.random.uniform(0, 16,(3,3))
B = np.random.uniform(30, 46,(3,3))

A_flat=A.flatten()
B_flat=B.flatten()

x =np.arange(len(A_flat))

#b. Plot these vectors on the same graph using matplotlib to compare how they look.
plt.plot(x,A_flat,marker='o',linestyle='-',label="Matrix A")
plt.plot(x,B_flat,marker='s',linestyle='-',label="Matrix B")

#c. Label the axes appropriately and give the plot a title.
plt.xlabel("index")
plt.ylabel("values")
plt.title("Line graph of flattened matrices A and B")
plt.legend()
plt.grid(True)


print("Matrix A:\n",A)
print("Matrix B:\n",B)
print("Matrix A flat:\n",A_flat)
print("Matrix B flat:\n",B_flat)
plt.show()
```
