A customer service center studies how long customers wait before being served.
They suspect:
 On regular days, customers arrive randomly and wait times are uniformly distributed
between 2 and 12 minutes.
 On busy days, wait times follow a normal distribution with an average (mean) of 8
minutes and a standard deviation of 2 minutes.
Your task is to simulate and analyze both situations using Python.

Problem Statement
Write a Python program to simulate and compare customer wait times for regular and busy days.

Tasks
🔹 Part 1 — Data Generation
1. Generate 3,000 wait times using a uniform distribution between 2 and 12 minutes.
2. Generate 3,000 wait times using a normal distribution with:
o Mean = 8
o Std Dev = 2
3. Ensure no negative wait times in the normal distribution (replace negatives with 0).

🔹 Part 2 — Visualization
1. Plot histograms for both distributions.
2. Show them:
o In the same figure (two subplots), OR
o Overlaid with transparency.

🔹 Part 3 — Statistical Analysis
For both datasets calculate:
 Mean
 Median
 Standard deviation
 Maximum wait time

# Necessary Libraries

```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
"""
Tasks 🔹 Part 1 — Data Generation

    Generate 3,000 wait times using a uniform distribution between 2 and 12 minutes.
    Generate 3,000 wait times using a normal distribution with:

    Mean = 8
    Std Dev = 2

    Ensure no negative wait times in the normal distribution (replace negatives with 0).


"""
#Taking my Last 2 digit number for seed
np.random.seed(65)

#Generating 3,000 wait times using a uniform distribution between 2 and 12 minutes.
regl_day = np.random.uniform(2, 12, 3000)


#Generate 3,000 wait times using a normal distribution with Mean = 8 and Std Dev = 2
busy_day = np.random.normal(8, 2, 3000)

#Here , Ensuring no negative wait times in the normal distribution (replace negatives with 0).
busy_day = np.where(busy_day< 0, 0,busy_day)
```

```python
"""
🔹 Part 2 — Visualization
1. Plot histograms for both distributions.
2. Show them:
o In the same figure (two subplots), OR
o Overlaid with transparency.


"""
#

plt.figure(figsize=(16,5))

# Regular Day
plt.subplot(1,2,1)
sns.histplot(regl_day, bins=30)
plt.title("Regular Day in Uniform Distribution")
plt.xlabel("Wait Time in Min")
plt.ylabel("Frequency")

# Busy Day
plt.subplot(1,2,2)
sns.histplot(busy_day, bins=30)
plt.title("Busy Day in Normal Distribution")
plt.xlabel("Wait Time in Min")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
"""
Part 3 — Statistical Analysis
For both datasets calculate:
 Mean
 Median
 Standard deviation
 Maximum wait time

"""
#Output of Regular Day
print("Regular Day Statistics:")
print("Mean:", np.mean(regl_day))
print("Median:", np.median(regl_day))
print("Standard Deviation:", np.std(regl_day))
print("Maximum:", np.max(regl_day))

#Output of Busy Day

print("\nBusy Day Statistics:")
print("Mean:", np.mean(busy_day))
print("Median:", np.median(busy_day))
print("Standard Deviation:", np.std(busy_day))
print("Maximum:", np.max(busy_day))
```
