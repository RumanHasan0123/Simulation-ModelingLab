```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
"""
TASK 1. Create Your Personal Matrix Let:
● d1 = last digit of your student ID
● d2 = second last digit of your student ID
Create the matrix : A = np.array([[d1+2, d2+1],[2*d1, d2+2]])
"""

# Last digit of my ID is 5 and 2nd last is 6
d1 = 5
d2 = 6

#Create the matrix : A = np.array([[d1+2, d2+1],[2*d1, d2+2]])
A = np.array([[d1+2, d2+1],[2*d1, d2+2]])

# The Matrix A
print("Matrix A:\n",A)
```

```python
"""
2. Analyze Matrix A
Using NumPy, compute:
● Shape of A
● Determinant
● Rank
● Eigenvalues
● Inverse (only if determinant ≠ 0)
"""


# dhape of Matrix A
size_A = A.shape

# determinant of Matrix A
det_A = np.linalg.det(A)

# rank of Matrix A
rank_A = np.linalg.matrix_rank(A)

# eigenvalues of Matrix A
eival_A  = np.linalg.eigvals(A)

# Inverse of Matrix A
if np.linalg.det(A) != 0:
    inv_A = np.linalg.inv(A)
else:
    inv_A = "[***ERROR:001***]Matrix A is singular, cannot compute inverse"

# Output
print("shape of Matrix A:\n", size_A)
print("determinant of matrix A:\n", det_A)
print("rank of Matrix A:\n", rank_A)
print("eigenvalues of Matrix A:\n", eival_A)
print("inverse of Matrix A:", inv_A)
```

```python
"""
3. Value Change Experiment
Create a new matrix B by changing exactly one value in A:
● Add +1 to one element of your choice
● You must clearly state which element you changed
"""

# Matrix B
# 1st element has been changed to  d1+3 from d1+2
B = np.array([[d1+3, d2+1],[2*d1, d2+2]])

# The Matrix B
print("Matrix B:\n",B,"\n")
```

```python
"""
4. Re-analyze Matrix B
Compute the same properties as in Step 2.
"""


# dhape of Matrix B
size_B = B.shape

# determinant of Matrix B
det_B = np.linalg.det(B)

# rank of Matrix B
rank_B = np.linalg.matrix_rank(B)

# eigenvalues of Matrix B
eival_B  = np.linalg.eigvals(B)

# Inverse of Matrix B
if np.linalg.det(B) != 0:
    inv_B = np.linalg.inv(B)
else:
    inv_B = "[***ERROR:002***]Matrix B is singular, cannot compute inverse"

# Output
print("shape of Matrix B:\n", size_B)
print("determinant of matrix B:\n", det_B)
print("rank of Matrix B:\n", rank_B)
print("eigenvalues of Matrix B:\n", eival_B)
print("inverse of Matrix B:", inv_B)
```
